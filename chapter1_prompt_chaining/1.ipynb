```python
import os 
from langchain_core.prompts import ChatMessagePromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_ollama import OllamaLLM
from langchain_core.runnables import RunnablePassthrough
```

```python
def get_llm():
    return OllamaLLM(
        model="tinyllama:1.1b",
        temperature=0.2
    )
```

```python
#prompt 1

prompt_extract = ChatMessagePromptTemplate.from_template(
    'Extract the technical specifications from the following text:\n\n{text_input}',
    role='user'
)
```

```python
#prompt 2

prompt_transform=ChatMessagePromptTemplate.from_template(
    "Transform the follwing specification into a JSON object with 'cpu','memory', and 'storage' as keys:\n\n{specifications}",
    role='user'
)
```

```python
llm=get_llm()
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_extract = ChatPromptTemplate.from_messages([prompt_extract])
prompt_transform = ChatPromptTemplate.from_messages([prompt_transform])

extraction_chain = prompt_extract | llm | StrOutputParser()
```

```python
full_chain=({"specifications":extraction_chain}|prompt_transform|llm|StrOutputParser())
```

```python
#run the chain
input_text='the new laptop model features a 3.5 GHz octa-core processor ,16GB of RAM, and a 1TB NVMe SSD'
final_result = full_chain.invoke({'text_input': input_text})
```
