```python
import os 
from langchain_core.prompts import ChatMessagePromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_ollama import OllamaLLM
from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import ChatPromptTemplate
```

```python
def get_llm():
    return OllamaLLM(
        model="tinyllama:1.1b",
        temperature=0.2
    )
```

```python
llm=get_llm()
```

# Simple Summarization Chain

```python
summary_prompt=ChatPromptTemplate.from_template(
    'summarize the following text:\n {text}'
)

chain=summary_prompt|llm

chain = summary_prompt | llm
result = chain.invoke({'text':'Langchain is a framework for building LLM applications'})

print(result)
```

# Summarization -> Bullet Points

```python
summary_prompt=ChatPromptTemplate.from_template(
    'summarize this:\n{text}'
)

bullet_prompt=ChatPromptTemplate.from_template(
    'Convert this summary in 5 bullet prompt:\n{summary}'
)

summary_chain=summary_prompt|llm|StrOutputParser()

chain=(
    {'summary':summary_chain}
    |bullet_prompt
    |llm
    |StrOutputParser()
)

result=chain.invoke({'text':'Langchain provides chains,agents,memory,retrievers,orchestrations....'})
print(result)
```

# Blog Idea ->outline->Blog

```python
idea_prompt=ChatPromptTemplate.from_template(
    "Generate 5 blog idea about {topic}"
)

outline_prompt=ChatPromptTemplate.from_template(
    'Create a outline from:\n {ideas}'
)

blog_prompt=ChatPromptTemplate.from_template(
    'Write a blog using this outline:\n {outline}'
)

idea_chain=idea_prompt|llm|StrOutputParser()

outline_chain=(
    {'ideas':idea_chain}
    | outline_prompt
    |llm
    |StrOutputParser()
)

blog_chain=(
    {'outline':outline_chain}
    |blog_prompt
    |llm
    |StrOutputParser()
)

result=blog_chain.invoke({'topic':'AI Agents'})

print(result)
```

# Translation chain
#### English -> Hindi -> Simple Hindi

```python
translate_prompt=ChatPromptTemplate.from_template(
    'Translate into French:\n{text}'
)

simplify_prompt=ChatPromptTemplate.from_template(
    'Simplify this French text:\{text}'
)

translation_chain=translate_prompt|llm|StrOutputParser()

chain=(
    {'text':translation_chain}
    |simplify_prompt
    |llm
    |StrOutputParser()
)

result=chain.invoke({'text':'Artificial Intelligence is transforming Industries'})
print(result)
```

# Email Generator

Purpose ->Draft -> Polish

```python
draft_prompt=ChatPromptTemplate.from_template(
    'Write an email about:\n{purpose}'
)

polish_prompt=ChatPromptTemplate.from_template(
    'Make this email professional:\n{email}'
)

draft_chain=draft_prompt|llm|StrOutputParser()

chain=(
    {'email':draft_chain}
    |polish_prompt
    |llm
    |StrOutputParser()
)

result=chain.invoke({'purpose':'sick leave'})
print(result)
```
